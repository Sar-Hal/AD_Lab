```python
import numpy as np

# 1. Create an array of 10 random values between 0 and 30
# (You can change 10 to any size 'X' you need)
original_array = np.random.uniform(0, 30, 10)

# 2. Define the new range
new_min = 12
new_max = 30

# 3. Normalize the array
# Formula: (x - min) / (max - min) * (new_max - new_min) + new_min
normalized_array = ((original_array - original_array.min()) / 
                    (original_array.max() - original_array.min()) * (new_max - new_min) + new_min)

print("Original Array:\n", original_array)
print("\nNormalized Array (between 12 and 30):\n", normalized_array)
```
