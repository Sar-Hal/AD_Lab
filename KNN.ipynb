```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("data.csv")

# Query point (change these values if needed)
q_weight = 5
q_height = 33

# Compute Euclidean distance
df["Distance"] = np.sqrt(
    (df["Weight"] - q_weight)**2 +
    (df["Height"] - q_height)**2
)

# Sort by distance
df_sorted = df.sort_values("Distance")

# Take k = 5 nearest neighbors
k = 5
knn = df_sorted.head(k)

print("5 Nearest Neighbors:\n")
print(knn)

# Majority voting
prediction = knn["Species"].mode()[0]

print("\nPredicted Class:", prediction)
```
