```python
#1 Feature Selection
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('Mall_Customers.csv')

# Select relevant features
X = df[['Annual Income (k$)', 'Spending Score (1-100)','Age']] 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=5, random_state=42) # Initialize KMeans with 5 clusters and a fixed random state
df['Cluster'] = kmeans.fit_predict(X_scaled) # Add cluster labels to original DataFrame
centroids = kmeans.cluster_centers_ # Get centroids in original scale
```

```python
import plotly.graph_objects as go

# Create 3D scatter for data points
scatter = go.Scatter3d(
    x=X_scaled[:,0],   # Income
    y=X_scaled[:,1],   # Spending
    z=X_scaled[:,2],   # Age
    mode='markers',
    marker=dict(
        size=5,
        color=df['Cluster'],
        colorscale='Viridis',
        opacity=0.8
    ),
    name='Customers'
)

# Create 3D scatter for centroids
centroid_plot = go.Scatter3d(
    x=centroids[:,0],
    y=centroids[:,1],
    z=centroids[:,2],
    mode='markers',
    marker=dict(
        size=12,
        color='red',
        symbol='x'
    ),
    name='Centroids'
)

# Create figure
fig = go.Figure(data=[scatter, centroid_plot])

fig.update_layout(
    title='Interactive 3D K-Means Clustering (Income, Spending, Age)',
    scene=dict(
        xaxis_title='Scaled Annual Income',
        yaxis_title='Scaled Spending Score',
        zaxis_title='Scaled Age'
    ),
    width=900,
    height=700
)

fig.show()
```

```python
cluster_analysis = df.groupby('Cluster')[
    ['Annual Income (k$)', 'Spending Score (1-100)', 'Age']
].mean()

print(cluster_analysis)
```
