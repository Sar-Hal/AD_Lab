```python
# ==========================================
# 1️⃣ Import Libraries
# ==========================================
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score


# ==========================================
# 2️⃣ Load Dataset
# ==========================================
df = pd.read_csv('Mall_Customers.csv')


# ==========================================
# FUNCTION TO RUN KMEANS + ANALYSIS
# ==========================================
def run_kmeans_analysis(X, df_original, model_name):
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    kmeans = KMeans(n_clusters=5, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)
    
    df_temp = df_original.copy()
    df_temp['Cluster'] = clusters
    
    print("\n=====================================")
    print(f"📊 Analysis for: {model_name}")
    print("=====================================")
    
    # Cluster Means
    print("\nCluster Means:")
    print(df_temp.groupby('Cluster').mean(numeric_only=True))
    
    # Cluster Sizes
    print("\nCluster Sizes:")
    print(df_temp['Cluster'].value_counts())
    
    # Inertia
    print("\nInertia (WCSS):", kmeans.inertia_)
    
    # Silhouette Score
    sil_score = silhouette_score(X_scaled, clusters)
    print("Silhouette Score:", sil_score)
    
    return sil_score, kmeans.inertia_


# ==========================================
# 3️⃣ MODEL 1 — 2D (Income + Spending)
# ==========================================
X_2D = df[['Annual Income (k$)', 'Spending Score (1-100)']]
sil_2d, inertia_2d = run_kmeans_analysis(X_2D, df, "2D Model (Income + Spending)")


# ==========================================
# 4️⃣ MODEL 2 — 3D (Income + Spending + Age)
# ==========================================
X_3D = df[['Annual Income (k$)', 
           'Spending Score (1-100)', 
           'Age']]
sil_3d, inertia_3d = run_kmeans_analysis(X_3D, df, "3D Model (Income + Spending + Age)")


# ==========================================
# 5️⃣ MODEL 3 — All Features (with Gender)
# ==========================================
df_encoded = pd.get_dummies(df, columns=['Genre'], drop_first=True)
X_all = df_encoded.drop(columns=['CustomerID'])

sil_all, inertia_all = run_kmeans_analysis(X_all, df_encoded, "All Features Model")


# ==========================================
# 6️⃣ Final Comparison Summary
# ==========================================
print("\n=====================================")
print("🔥 FINAL MODEL COMPARISON")
print("=====================================")

print(f"2D Model  → Silhouette: {sil_2d:.4f}, Inertia: {inertia_2d:.2f}")
print(f"3D Model  → Silhouette: {sil_3d:.4f}, Inertia: {inertia_3d:.2f}")
print(f"All Feat  → Silhouette: {sil_all:.4f}, Inertia: {inertia_all:.2f}")
```
